## Association Rule Mining

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
te = TransactionEncoder()
trans_data = te.fit_transform(unique_data['product']).astype('int')
```

```python
trans_data
```

```python
te.columns_
```

```python
encoded_data = pd.DataFrame(trans_data, columns=te.columns_)
encoded_data
```

```python
aprio = apriori(encoded_data, min_support= 0.3, use_colnames= True)
aprio
```

```python
association_rules(aprio, metric='confidence')
```

```python
aprio = apriori(encoded_data, min_support= 0.1, use_colnames= True)
print(aprio)
association_rules(aprio, metric='confidence')
```

## --------------------------------------------------------------------------------------

# FP Growth 

```python
from mlxtend.frequent_patterns import fpgrowth

fpg = fpgrowth(encoded_data, min_support= 0.3, use_colnames= True)
print(fpg)
association_rules(fpg, metric= 'confidence')
```

```python
from surprise import Dataset, Reader, SVD
from surprise.model_selection import train_test_split
from itertools import chain
```

```python
#  user ; item ; rating ; [timestamp]
reader = Reader(line_format= "user item rating", rating_scale= (0,5))
```

```python
# Null Perce

df.isnull().mean()*100
```

```python
df.fillna(0, inplace= True)
```

```python
data_for_sur = Dataset.load_from_df(df[['CustomerId', 'ProductID' , 'Rating']], reader)
```

```python
train_data, test_data = train_test_split(data_for_sur, test_size= 0.2)
```

```python
# Building model for the personalized recommendation
model = SVD()
```

```python
model.fit(train_data)
```

```python
model.predict(uid= 11891, iid= 701174210346869720)
```

```python
def generate_recomm(cust_id, number_of_rec):
    product_overall = df['ProductID'].unique()
    product_bought = df[df['CustomerId'] == cust_id]['ProductID'].values
    prod_used_for_predicct = [ i for i in product_overall if i not in product_bought ]
    predict_recomm = [ (prod,model.predict(cust_id,prod).est) for prod in prod_used_for_predicct ]
    output_recom = [ df[df['ProductID'] == i_d]['Product'].values[0] for i_d,a in predict_recomm[:number_of_rec] ]
    out_list = list(chain.from_iterable(output_recom))
    out_list = list(set(out_list))
    return out_list
```
